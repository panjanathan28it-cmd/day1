```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
doc1=nlp(u"Iam a runner running in a race because I love to run since I ran today")

for token in doc1:
  print(token.text, '\t', token.pos_, '\t', token.lemma, '\t', token.lemma_)
```

```python
def show_lemmas(text):
  for token in text:
    print(f'{token.text:{12}}  {token.pos_:{6}}  {token.lemma:<{22}}  {token.lemma_}')
```

```python
doc2 = nlp(u"iam saw eighteen mice today!")

show_lemmas(doc2)
```

```python
doc3=nlp(u"iam meeting tomorrow at the meeting.")

show_lemmas(doc3)
```
