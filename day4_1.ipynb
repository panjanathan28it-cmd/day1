```python
def vectorize_stories(data, word_index=tokenizer.word_index, max_story_len=max_story_len, max_question_len=max_question_len):
  X = []
  Xq = []
  Y = []

  # Add <unk> to word_index if it doesn't exist, and get its index
  if '<unk>' not in word_index:
    word_index['<unk>'] = len(word_index) + 1
  unk_idx = word_index['<unk>']

  for story, query, answer in data:
    # Vectorize story
    story_vec = [word_index.get(word.lower(), unk_idx) for word in story]
    # Vectorize query
    query_vec = [word_index.get(word.lower(), unk_idx) for word in query]

    # One-hot encode answer
    answer_vec = np.zeros(len(word_index) + 1)
    answer_vec[word_index[answer]] = 1

    X.append(story_vec)
    Xq.append(query_vec)
    Y.append(answer_vec)

  return (pad_sequences(X, maxlen=max_story_len), pad_sequences(Xq, maxlen=max_question_len), np.array(Y))
```
